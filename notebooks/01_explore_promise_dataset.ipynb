```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the JM1 dataset
df = pd.read_csv('../data/raw/jm1.csv')
```

```python
# Basic info
print("Dataset shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())
print("\nData types:")
print(df.dtypes)
print("\nMissing values:")
print(df.isnull().sum())
```

```python
# Target distribution (defects)
print("\nDefect distribution:")
print(df['defects'].value_counts(normalize=True))
```

```python
# Plot class imbalance
plt.figure(figsize=(6, 4))
sns.countplot(x='defects', data=df)
plt.title('Defect Distribution in JM1 Dataset (True = defective module)')
plt.xlabel('Defective (True/False)')
plt.ylabel('Count')
plt.show()
```

```python
# Quick look at some metrics
print("\nDescriptive statistics for key metrics:")
print(df[['loc', 'v(g)', 'ev(g)', 'iv(g)', 'n', 'v', 'l', 'd', 'i']].describe())
```
